```python
import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
from statistics import mean
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
def calculateSentiment(postId):
    r_post = requests.get(url=f"https://hacker-news.firebaseio.com/v0/item/{postId}.json")
    post = r_post.json()
    
    comments = []
    for item in post["kids"]:
        r_comment = requests.get(url=f"https://hacker-news.firebaseio.com/v0/item/{item}.json")
        comments.append(r_comment.json())
        
    compoundScores = []
    for comment in comments:
        compoundScore.append(sid.polarity_scores(comment["text"])['compound'])
        
    return mean(compoundScore)
```

```python
score = calculateSentiment(21955234)
```
